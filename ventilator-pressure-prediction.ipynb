```python
splits = RandomSplitter(valid_pct =0.2,seed = 42)(range_of(train_df))
#splits = RandomSplitter(valid_pct=0.2)(range_of(df))
```

```python
splits
```

```python
cont_names = [col for col in train_df if train_df[col].dtype!="O"]
cont_names
```

```python
cont_names.remove("pressure")
```

```python
cont_names
```

```python
to = TabularPandas(train_df,
                   procs = [Normalize],
                   cont_names = cont_names,
                   y_names = "pressure",
                   splits = splits)
```

```python
to.xs.iloc[:5]
```

```python
dls = to.dataloaders(bs=64)
```

```python
learn = tabular_learner(dls, metrics=mae)
```

```python
learn
```

```python
dls.valid.show_batch()
```

```python
learn.fit_one_cycle(1)
```

```python
learn.show_results()
```

```python
test_df
```

```python
dl = learn.dls.test_dl(test_df)
```

```python
dl
```

```python
pred = learn.get_preds(dl=dl)
```

```python
pred
```

```python
preasure = np.array(pred)
```
